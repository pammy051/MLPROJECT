```python
print('skewes values')


skewval_eachRow=df.skew(axis=1)
print(skewval_eachRow)
print('\n\nkurtosis values')

kurto=df.kurtosis(axis=1)
print(kurto)

print('\n\n skew with na include')
skewed_df_nainclude=df.skew(axis = 0, skipna = True)
print(skewed_df_nainclude)
```

```python
from scipy import stats
import numpy as np
sample_df=pd.DataFrame(df)
print(sample_df)
print('The null hypothesis for this test is that the data is a sample from a normal distribution, so a p-value less than 0.05 indicates significant skewness')

z = np.abs(stats.zscore(sample_df))   # get the z-score of every value with respect to their columns
print(z)
```
